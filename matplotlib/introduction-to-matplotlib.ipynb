## Matplotlib
  - python library for detailed graphs 

### pyplot
  - is a matplotlib module.
  - Creates figure, labels, and plot area


```python
import matplotlib.pyplot  as plt
```

### plot()
 - Draws points with default connecting lines
 `syntax: plot(x axis, y axis)`

 <p><strong>e.g. draw line in diagram</strong></p>

```python
# Basic line graph using plot()
plt.plot([1,3,2], [2,4,2])

#show plot & hide default title
plt.show()
```

<p><strong>Default X points values 0,1,2,3....</strong></p>

```python
import numpy as np

ypoints = np.array([2, 4, 1, 5])

#plotting without x-points
plt.figure()

plt.plot(ypoints) # y axis

plt.show()
```

### Format Strings
<p><strong>marker |line | color</strong></p>
 - optional 3rd argument in plot() for 
 <strong>'marker, 'line type' and 'color'</strong>

### Line Color Change
'color' or 'c' use shortcut string
- <strong>ways to specify colors:-</strong>
   
       Named COlors: 'red', 'green', 'blue' 
 
       Short Color codes: 'r' for red, 'g' for green, 'b' for blue

       RGB & Hexadecimal Color Codes




```python
plt.plot(ypoints, 'g')
plt.show()
```

### Change Line Style
 'linestyle' or use shortcut string
 `eg: '--', '-','.', ',', 'o', 'v', '^', '<', '>', '+', '=', '|', '_', '*', ':' `

```python
plt.plot(ypoints, 'o')
plt.show()
```

```python
plt.plot(ypoints, '*')
plt.show()
```

### Multple Lines 
  - Plot x and y axis seperately
  

```python
x = [10,4,7,1]
y = [2,8,3,9]

plt.plot(x, marker = '^', color = 'g')
plt.plot(y, marker = 'D', color = 'r')

plt.show()
```

### Marker
- mark the data points on a line plot
`styles available: 'o', 's', 'D', 'd', 'x', 'X', 'P', 'p', 'H', 'h', 'v', '1', '2', '3', '4', '^', '>', '<', '.', '*', ',', '|', '_', '-', '+'`

```python
plt.plot(ypoints, marker = 'o', color = 'g')
plt.show()
```

### Marker Size


argument `markersize` or rhe shorter version `ms`

```python
plt.plot(ypoints, marker = '*', ms = 15)
plt.show()
```

### Marker Color

`markeredgecolor` or `mec` to set the color of the edge `markerfacecolor` or `mfc` to set the color inside the edge

```python
plt.plot(ypoints, marker = 'o', ms = 24, mec = 'r', mfc = 'g')
plt.show()
```

### Marker + Line + Color

```python
# marker = 'o', line =':' color green 'g'

plt.plot(ypoints, 'o:g')
plt.show()
```

### Labels and Title
xlabels() and ylabels() functions to set label for x and y-axis title() function to set a title

```python
plt.title('Cen-Title')
plt.xlabel('Cen-x-axis')
plt.ylabel('Cen-y-axis')

plt.show()
```

### Font of Title and Labels
`fontdict` parameter

```python
font1 = {'family': 'serif', 'size': 25, 'color': 'blue'}
font2 = {'family': 'serif', 'size':20, 'color': 'green'}

#'loc' to position the title 

plt.title('Cen-tech',fontdict= font1, loc='right')
plt.xlabel('Cen-x-axis',fontdict= font2)
plt.ylabel('Cen-y-axis',fontdict= font2)

plt.show()
```

##### Create Mutliple Axes With Example

```python
# creaate data
x = [0, 1, 2, 3]
y = [3, 8, 1, 10]
```

```python
fig, ax = plt.subplots(1,2)

ax[0].plot(x, y)
ax[1].bar(x, y)

fig.suptitle('Figure of 2 Plot')

plt.show()
```

##### Alternate code


```python
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.plot(x, y)
ax2.bar(x, y)

fig.suptitle('fig, ax1 ax2')
plt.show()
```

#### Vertically stacked subplots

```python
fig, ax = plt.subplots(2)
# row = 0, column = 2

ax[0].plot(x, y)
ax[1].bar(x, y)

fig.suptitle('Vertically stacked Plots')
plt.show()
```
